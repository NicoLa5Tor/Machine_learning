```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.rand(1000) * 100 - 50
plt.plot(data)
plt.show()
```

El escalado de datos se usa para ajustar los valores de las caractaristicas de los datos, para que todas 
contribuyan de manera equitativa en el aprendizaje de la maquina. 

StandardScaler:

    Este método transforma los datos para que tengan una media de 0 y una desviación estándar de 1.
    Fórmula: z=(x−μ)/σ 

    x es el valor original.
    μ es la media de los datos.
    σ es la desviación estándar de los datos.

MinMaxScaler:

    Este método transforma los datos para que estén en un rango específico, generalmente [0, 1].
    Fórmula: 
    v = valor que queremos escalar
    m = valor minimo de los datos
    b = rango de los datos (lo que es igual a la resta del maximo de datos con el minomo del mismo)
    formula:
    (v-m) / (b)

RobustScaler:

    Este método es menos sensible a los valores atípicos. Escala los datos usando la mediana y el rango intercuartílico (IQR).
    Fórmula: xscaled=(x−mediana)IQRxscaled​=IQR(x−mediana)​
        xx es el valor original.
        medianamediana es la mediana de los datos.
        IQRIQR es el rango intercuartílico (Q3 - Q1).

Normalizer:

    Este método escala los datos para que cada muestra tenga una norma unitaria (longitud 1).
    Fórmula (para la norma L2): xscaled=x∥x∥2xscaled​=∥x∥2​x​
        xx es el vector de características de una muestra.
        ∥x∥2∥x∥2​ es la norma L2 del vector.

```python
# En esta ocacion se uso min max escaler
def min_max_scaler(data):
    min = np.min(data)
    max = np.max(data)
    return (data - min) / (max-min)
```

```python
#uso del metodo de min max escaler
min_max = min_max_scaler(data)
plt.plot(min_max)
plt.show()
```

Como podemos ver, la comparacion de las dos gráficas, en la que o se uso el metodo de escalado 
el valor va desde -40 hasta 40, pero con estos datos ya escalados el valor lo pone desde 0 hasta 1.

Esto hace que los datos sean más rapidos de tratar, y que los valores de cada dato tenga una diferencia equitativa enc omparación con los demás datos.

El comportamiento de los datos no cambia pero si la magnitud de estos, lo que los hace iguales pero mas pequeños, y si todos son peqeuños entonces no hay diferencia entre ellos que no se tuviese ya 


```python
#normalizado de datos por el metodo estandardScaler
def standardScaler(data):
    mdia = np.mean(data)
    #desviacion estandar
    std = np.std(data)
    return (data - mdia) / std
```

```python
stand_scaler = standardScaler(data)
plt.plot(stand_scaler)
plt.show()
```
