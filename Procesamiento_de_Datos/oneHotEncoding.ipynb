One Hot Encoding

Es una tecnica de procesamiento de tatos usada para variables categoricas en representaciones numericas,
esto se hace para algoritmos de aprendizaje, dato que las variables de aprendizaje debens er numéricas, 
lo que hace esta tecnica es hacer eso, convertir las variables string o de categoría en variables numericas o legibles para un algoritmo de aprendizaje

```python
data = pd.DataFrame({
    "Ubicacion" : ['Norte','Sur','Oeste','Norte'],
    "Cargo" : ['Gerente','Asistente','Analista','Gerente'],
    "Departamento" : ['Ventas','Marketing','Ventas','RRHH']
})
data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oneInit = OneHotEncoder()
data_encoder = pd.DataFrame(oneInit.fit_transform(data).toarray(),columns=oneInit.get_feature_names_out(data.columns))
data_encoder
```

Con pandas tambien podemos hacer un one hot encoding solo que esta vez en vez de ceros y unos nos arrojará falso o verdadero si el dato tiene ese campo específico

```python
data_encoding_pandas = pd.get_dummies(data)
data_encoding_pandas
```
